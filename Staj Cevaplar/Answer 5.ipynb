```python
import pandas as pd
```

```python
data=pd.read_csv('country_vaccination_stats.csv')
data.head(10)
```

```python
for col in data.columns:
    print("Column:",col)
    print(data[col].value_counts())
```

```python
#eksik değerlerin sayısı
data.isnull().sum().sum()
```

```python
#nan değerlerin bulunduğu yerler
data.isnull().sum()
```

```python
print(data.isnull().sum().sort_values(ascending=False))
```

```python
#NaN değerleri bulunduğu ülkedeki en küçük değere göre doldurma işlemi
```

```python
data["daily_vaccinations"].fillna(data.groupby("country")["daily_vaccinations"].transform("min"), inplace=True)
print(data.isnull().sum().sort_values(ascending=False))
```

```python
import numpy as np
```

```python
data.replace(np.nan, 0, inplace=True)
```
